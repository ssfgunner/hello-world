# 爬取20年内的基金数据

### 生成日期列表

```python
import datetime
import time
today = '2020-12-15'
date_list = []
time_tuple = time.strptime(today,"%Y-%m-%d")
year,month,day = time_tuple[:3]
for i in range(365*20):
    date_list.append((datetime.date(year,month,day)-datetime.timedelta(days=i)).strftime("%Y%m%d"))
date_list[0:5]
```

```python
import pandas as pd
import time
i = 0
for date in date_list:
    try:
        pro = ts.pro_api()
        time.sleep(0.75)
        df = pro.fund_daily(ts_code = '510050.SH',trade_date = date,asset='FD')
        if i == 0:
            data = df
            i+=1
        else:
            i+=1
            data = pd.concat([data,df])
        if i%100 == 0:
            print('爬取第{}天数据，数据总长度为:{}'.format(date,len(data)))
    except Exception as err:
        print(err)
        date_list.insert(0,date)
        time.sleep(30)
```

```python
data
```

### 给数据增加标签列，1表示涨，0表示跌

```python
data_sort = data.sort_values(['ts_code','trade_date'],ascending=['False','False'])
```

```python
label = data_sort['open']<=data_sort['close']
data_sort['label'] = label
data_sort.head()
```

```python
data_sort.to_csv('20_year_FD.csv',index = False)
```

# 数据特征构建

#### 为数据添加SMA，EMA等特征

```python
import talib as ta

df = data_sort[data_sort['ts_code']=='510050.SH']
types=['SMA','EMA','WMA','DEMA','TEMA',
'TRIMA','KAMA','MAMA','T3']
df_ma=pd.DataFrame(df.close)
for index,feature in enumerate(types):
    df_ma[feature] = ta.MA(df.close,timeperiod=5,matype=index)
df_ma
```
